# Step 3: Chi-square tests

```python
import numpy as np
from scipy.stats import chi2_contingency

# --- Chi-Square Test for RESULT_STAT_BP (Target) ---
target_variable = 'RESULT_STAT_BP'
categorical_predictors = [
    'total_income', 'union_name', 'age_group', 'gender',
    'is_poor', 'had_stroke', 'diabetic',
    'profile_hypertensive', 'RESULT_STAT_PR'
]
alpha = 0.05
results = []
for predictor in categorical_predictors:
    temp_df = df[[predictor, target_variable]].dropna()
    contingency_table = pd.crosstab(temp_df[predictor], temp_df[target_variable])

    if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:
        chi2, p_value, dof, expected = chi2_contingency(contingency_table)
        assumption_met = (expected >= 5).all()

        results.append({
            'Predictor': predictor,
            'Test': 'Chi-Square',
            'Statistic': chi2,
            'p-value': p_value,
            'Degrees_of_freedom': dof,
            'Assumption Met': 'True' if assumption_met else 'False',
            'Contingency_Table': contingency_table
        })


# --- Print Results ---
print("\n--- Statistical Test Results ---")
for result in results:
    print("-" * 30)
    print(f"Predictor: {result['Predictor']}")
    print(f"Test: {result['Test']}")

    # Print "p < 0.001" if p-value is very small
    if result['p-value'] is not None and result['p-value'] < 0.001:
        p_value_str = "< 0.001"
    elif result['p-value'] is not None:
        p_value_str = f"{result['p-value']:.6f}"
    else:
        p_value_str = "None"

    print(f"Test Statistic: {result['Statistic']:.4f}")
    if 'Degrees_of_freedom' in result:
        print(f"Degrees of Freedom: {result['Degrees_of_freedom']}")
    print(f"P-value: {p_value_str}")
    print(f"Assumption Met: {result['Assumption Met']}")

    significant_status = 'Yes' if result['p-value'] is not None and result['p-value'] < alpha else 'No'
    print(f"Significant: {significant_status}")

```

Show Contingency Table Here

Step 4: Visualizations for significant features

```python
print("\nGenerating combined heatmap and stacked bar chart visualizations...")

for result in results:
    if result['p-value'] < 0.05:
        feature = result['Predictor']
        contingency_table = result['Contingency_Table']

        # Normalize contingency table by row (proportions)
        normalized_table = contingency_table.div(contingency_table.sum(axis=1), axis=0)

        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        # Heatmap on left
        sns.heatmap(
            normalized_table,
            annot=True,
            cmap='viridis',
            fmt='.2f',
            cbar_kws={'label': 'Proportion'},
            ax=axes[0]
        )
        axes[0].set_title(f'Proportion Heatmap: RESULT_STAT_BP by {feature}', fontsize=14, weight='bold')
        axes[0].set_xlabel('RESULT_STAT_BP')
        axes[0].set_ylabel(feature)
        axes[0].tick_params(axis='x', rotation=45)
        axes[0].tick_params(axis='y', rotation=0)

        # Stacked bar chart on right
        normalized_table.plot(
            kind='bar',
            stacked=True,
            ax=axes[1],
            colormap='tab20'
        )
        axes[1].set_title(f'100% Stacked Bar Chart: {feature} vs RESULT_STAT_BP', fontsize=14, weight='bold')
        axes[1].set_ylabel('Proportion')
        axes[1].set_xlabel(feature)
        axes[1].tick_params(axis='x', rotation=45)
        axes[1].legend(title='RESULT_STAT_BP', bbox_to_anchor=(1.05, 1), loc='upper left')

        plt.tight_layout()
        plt.show()




```

# Step 5 Predictive Modeling

A Random Forest Classifier was applied to predict “Severe High” blood pressure using features such as total income, gender, diabetic status, hypertension profile, and RESULT_STAT_PR. Categorical variables were encoded using Label Encoding, and the dataset was split into 70% training and 30% testing with stratified sampling to maintain class balance. The model, trained with 200 decision trees, achieved a test accuracy of 98%, indicating strong predictive performance. Feature importance analysis revealed that RESULT_STAT_PR, profile_hypertensive, and total_income were the most influential predictors.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Copy and encode categorical variables
df_encoded = df.copy()
label_cols = ['total_income', 'gender', 'diabetic', 'profile_hypertensive', 'RESULT_STAT_PR', 'age_group']
for col in label_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Create binary target
df_encoded['Severe_High'] = (df['RESULT_STAT_BP'] == 'Severe High').astype(int)

# Features and target
features = ['total_income', 'gender', 'diabetic', 'profile_hypertensive', 'RESULT_STAT_PR']
X = df_encoded[features]
y = df_encoded['Severe_High']

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Train Random Forest model
model = RandomForestClassifier(
    n_estimators=200,     # Number of trees
    max_depth=None,       # Let trees grow fully
    random_state=42
)
model.fit(X_train, y_train)

# Results
print(f"Test Accuracy: {model.score(X_test, y_test):.2f}")

# Feature importance
importances = dict(zip(features, model.feature_importances_))
print("Feature Importances:", importances)
```
